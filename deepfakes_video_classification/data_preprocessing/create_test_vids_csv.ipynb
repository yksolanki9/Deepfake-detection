```python
from os.path import join
from os import listdir

from random import shuffle

import pandas as pd
```

```python
dirpath = "/scratch/ysolanki/FaceForensics++/test"
test_paths = [join(dirpath, '0'), join(dirpath, '1')]
list_0 = [join(test_paths[0], x) for x in listdir(test_paths[0])]
list_1 = [join(test_paths[1], x) for x in listdir(test_paths[1])]
```

```python
vid_list = list_0 + list_1
shuffle(vid_list)
```

```python
vids_list = []
label = []
for video in vid_list:
    vids_list.append(video)
    label.append(video.split('/')[-2])
```

```python
print(vids_list)
```

```python
data = {
    'vids_list':vids_list,
    'label':label
}

df = pd.DataFrame(data)
df.to_csv('/scratch/ysolanki/SiameseAkash/deepfakes_video_classification/test_data/test_vids_label.csv', index=False)
```
